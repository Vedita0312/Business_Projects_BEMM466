```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Step 1: Define your tasks and timelines
tasks = [
    ("Task Name", "Start Date", "End Date"),
    ("Final Report", "2025-08-19", "2025-08-25"),
    ("Synthesis & Recommendations", "2025-08-19", "2025-08-25"),
    ("Survey Analysis & Draft", "2025-08-12", "2025-08-18"),
    ("Survey Data Cleaning", "2025-08-05", "2025-08-11"),
    ("Survey Data Collection", "2025-07-29", "2025-08-11"),
    ("Survey Setup & Launch", "2025-07-29", "2025-08-04"),
    ("Indian Survey Design", "2025-07-22", "2025-07-28"),
    ("Draft UK Data Report", "2025-07-15", "2025-07-28"),
    ("XGBoost Model + SHAP", "2025-07-02", "2025-07-21"),
    ("UK Data Cleaning & EDA", "2025-07-02", "2025-07-14"),
]

# Step 2: Load into a DataFrame
df = pd.DataFrame(tasks[1:], columns=["Task", "Start", "End"])
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = pd.to_datetime(df["End"])
df["Duration"] = df["End"] - df["Start"]

# Step 3: Plot the Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))
for i, task in enumerate(df.itertuples()):
    ax.barh(task.Task, task.Duration.days, left=task.Start, color="skyblue", edgecolor="black")

# Step 4: Formatting
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MO))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.xticks(rotation=45)
ax.set_xlabel("Date")
ax.set_title("Data-Driven Pathways to Pay Equity")
plt.tight_layout()
plt.show()
```
